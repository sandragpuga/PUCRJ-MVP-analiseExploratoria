### **Sobre o conjunto de dados**

Este conjunto de dados fornece uma viso geral abrangente de vrios fatores que afetam o desempenho dos alunos em exames. Inclui informaes sobre hbitos de estudo, frequncia, envolvimento dos pais e outros aspectos que influenciam o sucesso acadmico.

### **Atributos**


*   **Horas_Estudadas:**	Nmero de horas gastas estudando por semana.

*   **Presena:** 	Porcentagem de aulas assistidas.


*   **Envolvimento Parental:**	Nvel de envolvimento dos pais na educao do aluno (baixo, mdio, alto).

*   **Acesso_a_Recursos:**	Disponibilidade de recursos
Item da lista educacionais (Baixa, Mdia, Alta).
*   **Atividades_extracurriculares:**	Participao em atividades extracurriculares (Sim, No).

*   **Horas de sono:**	Nmero mdio de horas de sono por noite.
*   **Pontuaes_anteriores:**	Notas de exames anteriores.

*   **Nvel_de_motivao:**	Nvel de motivao do aluno (baixo, mdio, alto).

*   **Acesso  Internet:**	Disponibilidade de acesso  internet (Sim, No).

*   **Sesses de tutoria:**	Nmero de sesses de tutoria frequentadas por ms.

*   **Renda_Familiar:**	Nvel de renda familiar (baixo, mdio, alto).

* **Qualidade_do_professor**:Qualidade dos professores (Baixa, Mdia, Alta).

*    **Tipo_de_escola:**	Tipo de escola frequentada (Pblica, Privada).

* **Influncia dos pares:**Influncia dos pares no desempenho acadmico (Positiva, Neutra, Negativa).

*   **Atividade_fsica:**	Nmero mdio de horas de atividade fsica por semana.

* **Dificuldades de aprendizagem:**	Presena de dificuldades de aprendizagem (Sim, No).

* **Nvel_de_Educao_Parental:**	Maior nvel de escolaridade dos pais (Ensino Mdio, Faculdade, Ps-graduao).

* **Distncia_de_casa:**	Distncia de casa at a escola (Perto, Moderado, Longe).

*   **Gnero:** Gnero do aluno (masculino, feminino).

*   **Pontuao do exame:** Nota do exame final.
